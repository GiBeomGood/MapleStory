```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
last_week = [82737, 40427, 79225, 5962, 61539, 7362, 69817, 8076, 5034, 36883, 2033, 1269, 79969, 50612]
this_week = [85933, 39935, 80984, 5779, 61716, 7207, 69431, 7703, 4769, 35656, 1857, 1211, 71016, 43692]
names = ['스카니아', '베라', '루나', '제니스', '크로아', '유니온', '엘리시움', '이노시스', '레드', '오로라', '아케인', '노바', '리부트', '리부트2']
```

```python
data1 = pd.DataFrame(zip(names, last_week), columns=['월드', '유저 수'])
data1['주'] = '지난주'

data = pd.DataFrame(zip(names, this_week), columns=['월드', '유저 수'])
data['주'] = '이번 주'

data = pd.concat([data1, data], axis=0)
del data1
data
```

```python
fig, ax = plt.subplots(1, 1, dpi=100, figsize=(7, 5))
sns.barplot(data=data, x='월드', y='유저 수', hue='주')
ax.set_xticklabels(labels=ax.get_xticklabels(), rotation=90);
```
