```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('어빌 확률표.csv')
data.applymap(lambda x: x if type(x) != float else '' if x == 0 else f'{x:5.2%}')
```

```python
def unique_option_prob(target_index):
    indices_list = list(range(len(data)))
    indices_list.remove(target_index)

    prob_second_line = 0.15 * sum(
        data.iloc[index, 3] * data.iloc[target_index, 2] / (1 - data.iloc[index, 2]) \
        for index in indices_list
    )

    indices_list = [(i, j) for i in indices_list for j in indices_list if i != j]

    prob_third_line = 0.15 * sum(
        data.iloc[index1, 3] * \
        (0.15 * data.iloc[index2, 2] / (1 - data.iloc[index1, 2]) + 0.85 * data.iloc[index2, 1] / (1 - data.iloc[index1, 1])) * \
        (data.iloc[target_index, 2] / (1 - data.iloc[index1, 2] - data.iloc[index2, 2])) \
        for (index1, index2) in indices_list
    )

    target_prob = prob_second_line + prob_third_line
    return target_prob
```

```python
target_index = 8
expectation = 4000 / unique_option_prob(target_index)
print(f'옵션 [{data.iloc[target_index, 0]:>27s}] 기대값: {int(expectation):9,}')
```
